```python
import pandas as pd
```

```python
df=pd.read_csv('strain.csv',names=['Time','Rad Flow',' Fpv Close','Fpv Open','High ','Bypass','Bpv Close','Bpv Open','class'])
```

```python
df
```

```python
features=df.columns[:-1].tolist()
```

```python
features
```

```python
x_train=df[features].values
y_train=df['class'].values
```

```python
x_train
```

```python
y_train
```

```python
df2=pd.read_csv('stest.csv',names=['Time','Rad Flow',' Fpv Close','Fpv Open','High ','Bypass','Bpv Close','Bpv Open','class'])
```

```python
df2
```

```python
x_test=df2[features].values
y_test=df2['class'].values
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler()
sc.fit(x_train)
x_train_std=sc.transform(x_train) 
x_test_std=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import accuracy_score
accuracy=[] 
for i in range(1,11):   
    knn=KNeighborsClassifier(n_neighbors=i,p=2,metric='minkowski')   
    knn.fit(x_train_std,y_train)   
    y_pred_knn=knn.predict(x_test_std)   
    print('No of missclassified values for k={} is {}'.format(i,(y_test!=y_pred_knn).sum())) 
    print('Accuracy for K={} is {}'.format(i,accuracy_score(y_test,y_pred_knn)*100))    
    accuracy.append(accuracy_score(y_test,y_pred_knn)*100)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=100,random_state=0,penalty='l1')
lr.fit(x_train_std,y_train)
y_pred_lr=lr.predict(x_test_std)
print('Misclassified Samples: %d'%(y_test!=y_pred_lr).sum())
print('Accuracy : %f'%(accuracy_score(y_test,y_pred_lr)))
```

```python
from sklearn.metrics import confusion_matrix 
c =confusion_matrix(y_pred_lr, y_test)
print(c)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
'C':[10, 20, 25, 50, 100, 200, 500],
'penalty':['l1','l2']
}

gsrf = GridSearchCV(estimator = lr, param_grid = parameters)
gsrf.fit(x_train_std, y_train)
print(gsrf.best_score_)
print(gsrf.best_params_)
print(gsrf.best_estimator_)
y_pred_lr_ht=gsrf.predict(x_test_std)
```

```python
print('Accuracy : %f'%(accuracy_score(y_test,y_pred_lr_ht)))
```

```python
c =confusion_matrix(y_pred_lr_ht, y_test)
print(c)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(criterion='entropy',n_estimators=10,random_state=1)
forest.fit(x_train_std,y_train)
y_pred_rf=forest.predict(x_test)
print('misclassified samples:%d'%(y_test!=y_pred_rf).sum())
print('Accuracy : %f'%(accuracy_score(y_test,y_pred_rf)))
```

```python
from sklearn.metrics import confusion_matrix 
c =confusion_matrix(y_pred_rf, y_test)
print(c)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'criterion':['gini', 'entropy'],
    'n_jobs':[-1],
    'n_estimators':[15, 20, 25],
    'max_depth':[2, 3, 4, 5, 10]
}

gsrf = GridSearchCV(estimator = forest, param_grid = parameters)
gsrf.fit(x_train_std, y_train)
print(gsrf.best_score_)
print(gsrf.best_params_)
print(gsrf.best_estimator_)
y_pred_rf_ht=gsrf.predict(x_test_std)
```

```python
print('Accuracy : %f'%(accuracy_score(y_test,y_pred_rf_ht)))
```

```python
c =confusion_matrix(y_pred_rf_ht, y_test)
print(c)
```
